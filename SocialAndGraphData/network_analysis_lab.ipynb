## Graph Measures

We can get the degree of all the nodes in the graph:

```python
nx.degree(G)
```

Or of a single node:

```python
G.degree('1')
```

We can also find the number of and show the connected components of the graph:

```python
nx.number_connected_components(G)
```

```python
list(nx.connected_components(G))
```

Other possible measures include the clustering coefficient, listing all the shortest paths, the betweenness centrality, or the PageRank measures:

```python
nx.clustering(G)
```

```python
nx.shortest_path(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.edge_betweenness_centrality(G)
```

```python
nx.pagerank(G)
plt.hist(list(nx.pagerank(G).values()))
plt.show()
```

## Plotting The Degree Distribution

We can plot the degree distribution of the graph using a histogram:

```python
diameter = nx.diameter(G)
print(diameter)
```

```python
average_distance = nx.average_shortest_path_length(G)
print(average_distance)

```

```python
deg_seq = [G.degree(x) for x in nx.nodes(G)]
plt.hist(deg_seq)
plt.title('Degree Histogram')
plt.xlabel('degree')
plt.ylabel('frequency')
plt.show()
```

```python
average_deg = float(sum(deg_seq))/float(len(deg_seq))
```

```python
p = average_deg/(len(G.nodes())-1)
Random_Version = nx.gnp_random_graph(len(G.nodes()), p)
nx.draw(Random_Version, with_labels=True)
plt.show()
plt.hist([Random_Version.degree(x) for x in nx.nodes(Random_Version)])
plt.title('Degree Histogram of Random Graph')
plt.xlabel('degree')
plt.ylabel('frequency') 
plt.show()
print("Average Degree of Original Graph:", average_deg)
print("Average Degree of Random Graph:", float(sum([Random_Version.degree(x) for x in nx.nodes(Random_Version)]))/float(len(Random_Version.nodes())))
print("Diameter of Original Graph:", diameter)
print("Diameter of Random Graph:", nx.diameter(Random_Version))
print("Average Distance of Original Graph:", average_distance)
print("Average Distance of Random Graph:", nx.average_shortest_path_length(Random_Version))
print("Clustering Coefficient of Original Graph:", nx.average_clustering(G))
print("Clustering Coefficient of Random Graph:", nx.average_clustering(Random_Version))
```

```python
# To count triangles one would need to focus on the neighbors of my neighbors being also my neighbor. Therefore, for two neighbors, i need to see if they are neighbors. This is similar to clustering coefficient calculation.
triangles = {}
for node in G.nodes():
    neighbors = list(G.neighbors(node))
    triangle_count = 0
    for i in range(len(neighbors)):
        for j in range(i + 1, len(neighbors)):
            if G.has_edge(neighbors[i], neighbors[j]):
                triangle_count += 1
    triangles[node] = triangle_count
print(triangles)

nx.algorithms.cluster.triangles(G)

print( "num triangles = ", sum(triangles.values())/3)
average_num_traingles = sum(triangles.values())/(3*len(Random_Version.nodes()))
print("average num triangles per node = ", average_num_traingles)
histogram_values = list(triangles.values())
plt.hist(histogram_values)
plt.title('Triangle Count Histogram')
plt.xlabel('Number of triangles a node is part of')
plt.ylabel('Frequency')
plt.show()
```
